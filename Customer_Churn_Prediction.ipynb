```python
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.sample(5)
```

```python
#drop customer id 

df.drop('customerID', axis= 'columns', inplace= True)
df.dtypes
```

```python
#convert total charges to integer 

df.TotalCharges.values
df.MonthlyCharges.values

pd.to_numeric(df.TotalCharges)
```

```python
#view error values i.e. values that have space in them 

pd.to_numeric(df.TotalCharges, errors='coerce').isnull()

# errors='coerce' adds n/a to the values that have space in them 
```

```python
# display rows with null values 

df[pd.to_numeric(df.TotalCharges, errors='coerce').isnull()]
```

```python
#remove rows where Total Charges= ''

df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.dtypes
```

```python
#after deleting empty rows we now convert the rest to integer 

df1.TotalCharges= pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtypes
```

```python
#find out unique values in each column 

for column in df:
  print(f'{column} : {df[column].unique()}') 
```

```python
#function to print columns which have object data type i.e. columns with yes and no value

def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes== 'object':
      print(f'{column} : {df[column].unique()}')   
```

```python
print_unique_col_values(df1)
```

```python
#data cleaning for processing 
#replace long strings containing no with No in dataframe

df1.replace('No internet service','No', inplace= True)
df1.replace('No phone service','No', inplace= True)

print_unique_col_values(df1)
```

```python
#convert yes and no to 1 and 0 for ease of ML models 

yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
                  'TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
  df1[col].replace({'Yes': 1, 'No': 0},inplace= True)
  
```

```python
for col in df1:
  print(f'{col}: {df1[col].unique()}')
```

```python
#similarly for male and female column 

df1['gender'].replace({'Female': 1, 'Male': 0},inplace= True)

df1['gender'].unique()
```

```python
#one hot encode the column 'Contract','InternetService','PaymentMethod' using pandas dummy function 

df2 = pd.get_dummies(data= df1, columns=['Contract','InternetService','PaymentMethod'])
df2.columns 
```

```python
df2.dtypes
```

```python
#scaling data 
#scale the columns 'tenure','MonthkyCharges','TotalCharges' using min max scaler
#min max scaler converts values in a given range (eg. 0-50) to 0-1

cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
scaler = MinMaxScaler()
df2[cols_to_scale]= scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
  print(f'{col}: {df2[col].unique()}')
```

```python
#train test and split of data 

X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
#create a neural network using Tensorflow

import tensorflow as tf 
from tensorflow import keras

#we give the number of neurons we want in our network 
#we have input layer, hidden layer(s) and output layer in neural network, 
#here hidden layer is (26),(15) and output is(1)

# model = keras.Sequential([
#     keras.layers.Dense(26, input_shape(26,), activation='relu'),
#     keras.layers.Dense(15, activation ='relu'), 
#     keras.layers.Dense(1, activation ='sigmoid')  
# ])

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

#adam is commonly used optimizer, loss is binary_crossentropy because output is binary 
# optimizers like Momentum, Nesterov, Adagrad, Adadelta, RMSProp, Adam and Nadam

model.compile(optimizer= 'adam',
              loss= 'binary_crossentropy',
              metrics=['accuracy'] )

model.fit(X_train, y_train, epochs=50)
```

```python
model.evaluate(X_test, y_test)
```

```python
#compare y predicted and y test

yp= model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
#convert 2d array to 1d

y_pred=[]
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:5]
```

```python
#print classification report which tells us the overall performance of our model 

from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,y_pred))
```

```python
#confusion matrix 

import seaborn as sn 
cm= tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot= True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
